```julia
using PyPlot
```

```julia
a1 = [0,0,1]
a2 = [0,1,0]
a3 = [1,0,0]
```

```julia
points = 3

x = zeros(points^3)
y = zeros(points^3)
z = zeros(points^3)

dom = linspace(-10,10,points)

for i in 1:points
    for l in 1:points
        for k in 1:points
            vec = dom[i]*a1 + dom[l]*a2 + dom[k]*a3
            x[i + (l-1)*points + (k-1)*points^2] = vec[1]
            y[i + (l-1)*points + (k-1)*points^2] = vec[2]
            z[i + (l-1)*points + (k-1)*points^2] = vec[3]
        end
    end
end
```

```julia
scatter3D(x,y,z,linewidth = 4)
axis("off")
savefig("cell.png")
show()
```

```julia
points = 40

x1 = zeros(points^3)
y1 = zeros(points^3)
z1 = zeros(points^3)

dom = linspace(-5,5,points)

for i in 1:points
    for l in 1:points
        for k in 1:points
            vec = dom[i]*a1 + dom[l]*a2 + dom[k]*a3
            x1[i + (l-1)*points + (k-1)*points^2] = vec[1]
            y1[i + (l-1)*points + (k-1)*points^2] = vec[2]
            z1[i + (l-1)*points + (k-1)*points^2] = vec[3]
        end
    end
end
```

```julia
scatter3D(x,y,z,linewidth = 4)
scatter3D(x1,y1,z1,linewidth = 1, c="red")
axis("off")
savefig("brill.png")
show()
```

```julia
plot(y1,z1, "ro", markersize=2)
plot(y,z,linewidth = 4 ,"bo", markersize=9)
axvline(5, color="r")
axvline(-5, color="r")
axhline(5, color="r")
axhline(-5, color="r")
axis("off")
axis("equal")
savefig("contour.png")
show()
```
